```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
features_df = pd.read_csv('../data/features data-set.csv')
```

```python
features_df.head()
```

```python
for feature in features_df.columns:
    print(f"The column {feature} has {features_df[feature].isna().sum()} nulls")
```

```python
# drop null values 
features_df.dropna(subset=["Unemployment", "CPI"], 
                   inplace=True)
```

```python
features_df.head()
for feature in features_df.columns:
    print(f"The column {feature} has {features_df[feature].isna().sum()} nulls")
```

```python
for column in ["CPI", 
               "Unemployment", 
               "Temperature",
               "Fuel_Price"]:
    plt.figure(figsize=(12, 7))
    plt.rcParams["figure.autolayout"] = True
    ax = sns.lineplot(data=features_df, 
                 x="Date", 
                 y=f"{column}")
    
    ax.tick_params(axis='x', 
                   rotation=75,
                   size=9,
                   pad=5)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
    plt.title(f"{column} by Date")
    plt.savefig(f"charts/{column}.png")
    plt.show()
    plt.close()
```
