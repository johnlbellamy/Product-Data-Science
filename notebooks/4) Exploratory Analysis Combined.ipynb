```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
store_data = pd.read_csv('../data/stores data-set.csv')
```

```python
sales_data = pd.read_csv('../data/sales data-set.csv')
```

```python
features_data = pd.read_csv('../data/features data-set.csv')
```

```python
store_and_sales = pd.merge(store_data, sales_data, left_on=["Store"], right_on=["Store"], how='left')
```

```python
store_and_sales
```

```python
store_and_features = pd.merge(store_and_sales,
                              features_data,
                           left_on=["Store", "Date"], 
                           right_on=["Store", "Date"], 
                           how='left')
```

```python
store_and_features
```

```python
store_and_features.fillna(0, inplace=True)
```

```python
store_and_features["Total Markdown"] = store_and_features["MarkDown1"] + store_and_features["MarkDown2"] + store_and_features["MarkDown3"] + store_and_features["MarkDown4"] + store_and_features["MarkDown5"]
```

```python
store_and_features.to_csv("../data/all_features.csv", index=False)
```
