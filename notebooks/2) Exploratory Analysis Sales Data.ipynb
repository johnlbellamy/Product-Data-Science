```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
features_df = pd.read_csv('../data/sales data-set.csv')
```

```python
features_df
```

```python
# Weekly Store Sales by Department
for feature in features_df.columns:
    print(f"The column {feature} has {features_df[feature].isna().sum()} nulls")
```

```python

    
plt.figure(figsize=(12, 7))
plt.rcParams["figure.autolayout"] = True
ax = sns.lineplot(data=features_df,
                  x="Date",
                  y=f"Weekly_Sales",
                  hue="Store")

ax.tick_params(axis='x',
               rotation=75,
               size=9,
               pad=5)
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))

plt.savefig(f"charts/sales_by_store.png")
plt.show()
plt.close()
```
